# Reading values of skeletal joints coordinates from .skeleton file
## Using python read() function to read the file


```python
import string
import numpy as np

def readskeletonfile(file):

    file = open(file,"r")
    framecount = int(file.readline())
    print (framecount)
#     jointlist = []
    coordinatelist = []
    for framenum in range(framecount):
        file.readline()
        file.readline()
        
        #converting the joint number string to integer
        jointsnum = int(file.readline())
        coordinates = []
        #looping through all 25 joints
        for joint in range(jointsnum-5):
            
            #getting the x y z coordinate from the files
            numbers = file.readline().split()
#             joint = Joints(numbers[0],numbers[1],numbers[2])
            coordinates.append(numbers[0])
            coordinates.append(numbers[1])
            coordinates.append(numbers[2])
            # 3D location of the joint j
#             jointlist.append(joint)
        
        file.readline()
        file.readline()
        file.readline()
        file.readline()
        file.readline()
        
        #converting list to numpy array
        numpyarray = np.array(coordinates, dtype=np.float32)
        if len(coordinatelist):
            #creating nested numpy array or 2D array
            coordinatelist = np.vstack((coordinatelist,numpyarray))
        else:
            coordinatelist = numpyarray
#     return jointlist
    return coordinatelist



```

```python
import h5py
import numpy as np
import os

f = h5py.File('D:/Major Project/pkummdtraindataset.h5', 'a')
# Creating  train dataset to store features and labels
train_dset = f.create_dataset('traindata', (3740, 300, 150), dtype='f')
trainl_dset = f.create_dataset('trainlabel', (3740, 8, 1), dtype='f')

# 2-brushing hair, 3-brushing teeth, 8-drink water, 13-handwaving, 17-jump up, 22-pick up, 32-salute, 33-sitting
actionlist = [2,3,8,13,17,22,32,33]
sample = 0
for actionfile in os.listdir("D:/Major Project/Skeleton/trainlabel/"):
    labelfile = open("D:/Major Project/Skeleton/trainlabel/" + actionfile,"r")
    value = labelfile.readline()
    while True:
        if (value == ''):
            break
        else:
            num = value.split(',')
            action = int(num[0])
            
            if action in actionlist:
                startframe = int(num[1])
                endframe = int(num[2])
                confidence = int(num[3])
                totalframe = endframe-startframe
                
                if (totalframe < 300 and confidence != 0):
                    
                    train_data = np.zeros((300, 150),dtype=np.float32)
                    target_data = np.zeros((8, 1), dtype=np.float32)
                    
                    if (action == 2):
                        target_data[0][0] = float(1)
                    elif (action == 3):
                        target_data[1][0] = float(1)
                    elif (action == 8):
                        target_data[2][0] = float(1)
                    elif (action == 13):
                        target_data[3][0] = float(1)
                    elif (action == 17):
                        target_data[4][0] = float(1)
                    elif (action == 22):
                        target_data[5][0] = float(1)
                    elif (action == 32):
                        target_data[6][0] = float(1)
                    elif (action == 33):
                        target_data[7][0] = float(1)
                        
                    datafile = open("D:/Major Project/Skeleton/pkuskeleton/" + actionfile,"r")
                    fr=0
                    for frame, line in enumerate(datafile):
                        if (frame >= startframe and frame <= endframe):
                            for i, value in enumerate(line.split()):
                                train_data[fr][i] = float(value)
                            fr+=1
                               
                    datafile.close()
                                    
                    train_dset[sample] = train_data
                    trainl_dset[sample] = target_data
                    sample+=1
                    
            value = labelfile.readline()
    labelfile.close()
    print("Feature from file:" + actionfile + " into h5 file done..")
f.close()
```
