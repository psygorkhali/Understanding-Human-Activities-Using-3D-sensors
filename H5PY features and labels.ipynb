# Store the whole dataset into h5py file
## Making the whole feature and label loading possible

```python
import h5py
import numpy as np
import os
import random

#creating lists of samples
samplelist = []
for actionfile in os.listdir("D:/Major Project/traindata/"):
    samplelist.append(actionfile)
print("Number of Samples")
print(len(samplelist))

f = h5py.File('D:/Major Project/totaltraindataset.h5', 'a')

# Creating  train dataset to store features and labels
train_dset = f.create_dataset('traindata', (10940, 300, 150), dtype='f')
trainl_dset = f.create_dataset('trainlabel', (10940, 8, 1), dtype='f')
    
for sample in range(7200):
    #storing each action/file into numpy array
    train_data = np.zeros((300, 150),dtype=np.float32)
    target_data = np.zeros((8, 1), dtype=np.float32)
    
    #choosing a random file from list
    actionfile = random.choice(samplelist)
     #enumerate(os.listdir("E:/nturgbd_skeletons/traindata")):
    file = open("D:/Major Project/traindata/" + actionfile,"r")
    #recognizing actions
    action = int(actionfile[18:20])
    if (action == 4):
        target_data[0][0] = float(1)
    elif (action == 3):
        target_data[1][0] = float(1)
    elif (action == 1):
        target_data[2][0] = float(1)
    elif (action == 23):
        target_data[3][0] = float(1)
    elif (action == 27):
        target_data[4][0] = float(1)
    elif (action == 6):
        target_data[5][0] = float(1)
    elif (action == 38):
        target_data[6][0] = float(1)
    elif (action == 8):
        target_data[7][0] = float(1)
           
    #read number of frame/ here frame number is the timesteps
    framecount = int(file.readline())
    #print("file no: "+ actionfile)
    for framenum in range(framecount):
        #we'll have total of 20joints with each x,y,z point so total of 20x3=60 feature vector
        #each frame has 60 features
        #print("frame number: "+str(framenum))
        feature = 0;
        #read the total skeleton number and ignore it
        bodycount = int(file.readline())
        for body in range(bodycount):
            #read the body info and ignore it
            file.readline()
            #read joint number and converting the joint number string to integer
            jointsnum = int(file.readline())
            if (body < 2 ):
                #looping through only 20 joints
                for joint in range(jointsnum):

                    #getting the x y z coordinate from the files and ignoring other information
                    numbers = file.readline().split()
                    train_data[framenum][feature] = numbers[0]
                    feature+=1
                    train_data[framenum][feature] = numbers[1]
                    feature+=1
                    train_data[framenum][feature] = numbers[2]
                    feature+=1
                #ignoring the remaining 5 joints
#                   for i in range(5):
#                       file.readline()
            else:
                for i in range(25):
                    file.readline()
                        
    print("Feature from file:" + actionfile + " into h5 file done..")                
    train_dset[sample] = train_data
    trainl_dset[sample] = target_data
    #removing the selected file
    samplelist.remove(actionfile)
    

#another dataset merging
# 2-brushing hair, 3-brushing teeth, 8-drink water, 13-handwaving, 17-jump up, 22-pick up, 32-salute, 33-sitting down
actionlist = [2,3,8,13,17,22,32,33]
sample = 7200
for actionfile in os.listdir("D:/Major Project/Skeleton/trainlabel/"):
    labelfile = open("D:/Major Project/Skeleton/trainlabel/" + actionfile,"r")
    value = labelfile.readline()
    while True:
        if (value == ''):
            break
        else:
            num = value.split(',')
            action = int(num[0])
            
            if action in actionlist:
                startframe = int(num[1])
                endframe = int(num[2])
                confidence = int(num[3])
                totalframe = endframe-startframe
                
                if (totalframe < 300 and confidence != 0):
                    
                    train_data = np.zeros((300, 150),dtype=np.float32)
                    target_data = np.zeros((8, 1), dtype=np.float32)
                    
                    if (action == 2):
                        target_data[0][0] = float(1)
                    elif (action == 3):
                        target_data[1][0] = float(1)
                    elif (action == 8):
                        target_data[2][0] = float(1)
                    elif (action == 13):
                        target_data[3][0] = float(1)
                    elif (action == 17):
                        target_data[4][0] = float(1)
                    elif (action == 22):
                        target_data[5][0] = float(1)
                    elif (action == 32):
                        target_data[6][0] = float(1)
                    elif (action == 33):
                        target_data[7][0] = float(1)
                        
                    datafile = open("D:/Major Project/Skeleton/pkuskeleton/" + actionfile,"r")
                    fr=0
                    for frame, line in enumerate(datafile):
                        if (frame >= startframe and frame <= endframe):
                            for i, value in enumerate(line.split()):
                                train_data[fr][i] = float(value)
                            fr+=1
                               
                    datafile.close()
                                    
                    train_dset[sample] = train_data
                    trainl_dset[sample] = target_data
                    sample+=1
                    
            value = labelfile.readline()
    labelfile.close()
    print("Feature from file:" + actionfile + " into h5 file done..")
    
f.close()
```
